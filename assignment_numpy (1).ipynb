* __Task 1__: Calculate the product of nonzero elements on the diagonal of a rectangular matrix.  
 For example, for X = np.array([[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]]) answer is 3.

```python
# your code
X = np.array([[1, 0, 1],
              [2, 0, 2],
              [3, 0, 3],
              [4, 4, 4]
            ])

result = 1

for i in range(min(len(X), len(X[0]))):
    if X[i][i] != 0:
      result*=X[i][i]
print(result)

d = np.diag(X)
result = d[d != 0].prod()
print(result)
```

* #### __Task 6__: CrunchieMunchies __*__

You work in the marketing department of a food company called MyCrunch, which is developing a new type of delicious, healthy cereal called **CrunchieMunchies**.

You want to show consumers how healthy your cereal is compared to other leading brands, so you've compiled nutrition data from several different competitors.

Your task is to use Numpy calculations to analyze this data and prove that your **СrunchieMunchies** - the healthiest choice for consumers.


```python
import numpy as np
```

1. View the cereal.csv file. This file contains calorie counts for various brands of cereal. Download the data from the file and save it as calorie_stats.

```python
calorie_stats = np.loadtxt("/data/cereal.csv", delimiter=",")
calorie_stats
```

2. One serving of CrunchieMunchies contains 60 calories. How much higher is the average calorie count of your competitors?

Store the answer in the variable average_calories and print the variable in the terminal

```python
# your code
average_calories = np.mean(calorie_stats) - 60
print(round(average_calories,2))
```

3. Does the average calorie count accurately reflect the distribution of the data set? Let's sort the data and see.

Sort the data and store the result in the calorie_stats_sorted variable. Print the sorted information.

```python
# your code
calorie_stats_sorted = np.sort(calorie_stats)
print(calorie_stats_sorted)
mean = np.mean(calorie_stats_sorted)
print(mean)
```

4. It looks like most values ​​are above the mean. Let's see if the median is the most accurate measure of the data set.

Calculate the median of the data set and save your answer in median_calories. Print the median so you can see how it compares to the mean.

```python
# your code
median_calories = np.median(calorie_stats)
print(median_calories)
print("Compare median to mean:",median_calories-mean)
```

5. While the median shows that at least half of our values ​​are over 100 calories, it would be more impressive to show that a significant portion of the competitors have higher calorie counts than CrunchieMunchies.

Calculate various percentiles and print them until you find the lowest percentile that exceeds 60 calories. Store this value in the nth_percentile variable.

```python
# your code
for i in range(101):
  percentile_value = np.percentile(calorie_stats,i)
  print(i,percentile_value)

  if percentile_value > 60:
    nth_percentile = i
    break
print("Percentile:",nth_percentile)
```

6. While the percentile shows us that most competitors have much higher calorie counts, it's an awkward concept to use in marketing materials.

Instead, let's calculate the percentage of cereals that contain more than 60 calories per serving. Save your answer in the more_calories variable and print it.

```python
# your code
more_calories = np.mean(calorie_stats>60)*100
print(more_calories)
```

7. This is a really high percentage. This will be very useful as we promote CrunchieMunchies. But one question is: how big are the differences in the dataset? Can we generalize that most cereals contain around 100 calories or is the range even wider?

Calculate the magnitude of the deviation by finding the standard deviation. Save your answer in calorie_std and print it on the terminal. How can we incorporate this value into our analysis?

```python
# your code
calorie_std = np.std(calorie_stats)
print(calorie_std)
```
